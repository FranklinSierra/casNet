# Reading and saving throught data

```python
#original data
folders = ["testA", "testB", "trainA", "trainB"]
for folder in folders:
    path_or = '../../../data/polyps/'+folder+'/'
    imgs = os.listdir(path_or)
    for img in imgs:
        image = Image.open(path_or+img)
        img2 = image.resize((512,512), Image.ANTIALIAS)
        file_name = '../cyclegan_tf/data/polyps/'+folder+'/'+img
        img2.save(file_name, quality=100, optimize=True)
print("finished!")
```
